```python
import numpy as np

import scikitplot as skplt
import matplotlib.pyplot as plt
from sklearn.externals import joblib
import LBP_Classifier 

# load dataset

X_training, y_training = LBP_Classifier.load_features_and_labels(forTraining=True)
y_training = LBP_Classifier.fix_shape_of_y(y_training)

X_testing, y_testing = LBP_Classifier.load_features_and_labels(forTraining=False)
y_testing = LBP_Classifier.fix_shape_of_y(y_testing)

X = np.vstack((X_training, X_testing))
y = np.hstack((y_training, y_testing))

# load classifier

svm_FASD = joblib.load('svm_FASD.pkl')
gnb_FASD = joblib.load('gnb_FASD.pkl')
knn_FASD = joblib.load('knn_FASD.pkl')
rf_FASD = joblib.load('rf_FASD.pkl')
lr_FASD = joblib.load('lr_FASD.pkl')
```

```python
# generate learning curve for svm
from sklearn import svm
svm = svm.SVC()
skplt.estimators.plot_learning_curve(svm, X, y, title='SVM Learning Curve')
plt.show()
```

```python
# generate learning curve for knn
from sklearn.neighbors import KNeighborsClassifier
k_nn = KNeighborsClassifier(n_neighbors=2)
skplt.estimators.plot_learning_curve(k_nn, X, y, title='KNN Learning Curve')
plt.show()
```

```python
# generate learning curve for gnb
from sklearn.naive_bayes import GaussianNB
g_nb = GaussianNB()
skplt.estimators.plot_learning_curve(g_nb, X, y, title='Gaussian Naive Bayes Learning Curve')
plt.show()
```

```python
# generate learning curve for rf
from sklearn.ensemble import RandomForestClassifier
rf_c = RandomForestClassifier()
skplt.estimators.plot_learning_curve(rf_c, X, y, title='Random Forest Learning Curve')
plt.show()
```

```python
# generate learning curve for lr
from sklearn.linear_model import LogisticRegression
l_r = LogisticRegression()
skplt.estimators.plot_learning_curve(l_r, X, y, title='Logistic Regression Learning Curve')
plt.show()
```

```python
# obtain the prediction and probability values from the svm_FASD classifier
y_pred_svm = svm_FASD.predict(X_testing)
y_prob_svm = svm_FASD.predict_proba(X_testing)

# obtain the prediction and probability values from the knn_FASD classifier
y_pred_knn = knn_FASD.predict(X_testing)
y_prob_knn = knn_FASD.predict_proba(X_testing)

# obtain the prediction and probability values from the gnb_FASD classifier
y_pred_gnb = gnb_FASD.predict(X_testing)
y_prob_gnb = gnb_FASD.predict_proba(X_testing)

# obtain the prediction and probability values from the rf_FASD classifier
y_pred_rf = rf_FASD.predict(X_testing)
y_prob_rf = rf_FASD.predict_proba(X_testing)

# obtain the prediction and probability values from the lr_FASD classifier
y_pred_lr = lr_FASD.predict(X_testing)
y_prob_lr = lr_FASD.predict_proba(X_testing)
```

```python
from sklearn import metrics
from scipy.optimize import brentq
from scipy.interpolate import interp1d
fpr, tpr, thresholds = metrics.roc_curve(y_testing, y_pred_svm)
eer = brentq(lambda x : 1. - x - interp1d(fpr, tpr)(x), 0., 1.)
#thresh = interp1d(fpr, thresholds)(eer)

mat = metrics.confusion_matrix(y_testing, y_pred_svm)
tp, tn, fp, fn = (mat[1,0], mat[0,0], mat[1,1], mat[0,1])
sensitivity = tp/tp+fn
specificity = tn/tn+fp
BCR = (sensitivity + specificity)/2
HTER = 1-BCR
print('Support Vector Machine metrics:')
print('Sensitivity:', sensitivity)
print('Specificity:', specificity)
print('Balanced Classification Rate:', BCR)
print('Half Total Error Rate', HTER)
print('Equal Error Rate:', eer)
```
